```python
# 1. 安装系统依赖 (约 2 分钟)
! apt-get update -qq && apt-get install -y -qq \
    python3-pip python3-dev build-essential \
    autoconf automake libtool libffi-dev \
    zlib1g-dev libncurses5-dev libncursesw5-dev \
    libssl-dev libsqlite3-dev \
    openjdk-17-jdk-headless zip unzip git \
    libltdl-dev cmake > /dev/null 2>&1
print('系统依赖安装完成')
```

```python
# 2. 安装 Buildozer (约 1 分钟)
! pip install --upgrade pip -q
! pip install buildozer cython -q
! buildozer version
print('Buildozer 安装完成')
```

```python
# 3. 克隆项目
! git clone https://github.com/Zhenfangan/soloist-carbin.git /content/soloist-carbin
! ls /content/soloist-carbin/app/main.py && echo '项目克隆完成'
```

```python
# 4. 构建 APK (约 20-30 分钟，首次需下载 SDK/NDK)
import os
os.chdir('/content/soloist-carbin')

# 清理可能存在的旧缓存
! rm -rf .buildozer/android/platform/python-for-android/ 2>/dev/null
! rm -rf .buildozer/android/platform/build-* 2>/dev/null

# yes 自动确认 root 警告
! yes | buildozer android debug
```

```python
# 5. 下载 APK 到本地
from google.colab import files
import glob

apks = glob.glob('/content/soloist-carbin/bin/*.apk')
if apks:
    for apk in apks:
        print(f'下载: {apk}')
        files.download(apk)
else:
    print('APK 未找到，查看上面的构建日志排查错误')
```
